```python
!pip install mtcnn
import os
import cv2
import numpy as np
import pandas as pd
from pathlib import Path
from tqdm import tqdm
from tensorflow.keras.applications import ResNet50, ResNet50V2
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import Adam
import tensorflow as tf
import matplotlib.pyplot as plt
import shutil
from mtcnn import MTCNN

# Set random seed for reproducibility
tf.random.set_seed(42)
np.random.seed(42)

# ========================================
# CONFIGURATION
# ========================================
ROOT = Path("/kaggle/input/new-ff-c23/Datasets")
TEST_VIDEO_DIR = Path("/kaggle/input/test-df-video")
TEST_IMAGE_DIR = Path("/kaggle/input/test-df-image")  # Add test image directory
IMG_SIZE = (224, 224)
BATCH_SIZE = 8
EPOCHS = 20
FRAMES_PER_VIDEO = None  # None = extract all frames, or set to number like 30
FACE_DETECTION_CONFIDENCE = 0.7  # Minimum confidence for face detection
FRAME_SKIP = 1  # Process every Nth frame (1 = all frames, 2 = every 2nd frame, etc.)

# ========================================
# DATA GENERATORS
# ========================================
print("ğŸ“‹ Setting up data generators...")

# Initialize MTCNN detector
print("ğŸ” Initializing MTCNN face detector...")
detector = MTCNN()

datagen = ImageDataGenerator(rescale=1./255)

train_gen = datagen.flow_from_directory(
    directory=ROOT / "train",
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode="binary",
    shuffle=True,
    classes=['fake', 'real']
)
print(f"Total training samples: {train_gen.samples}")

val_gen = datagen.flow_from_directory(
    directory=ROOT / "valid",
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode="binary",
    shuffle=False,
    classes=['fake', 'real']
)
print(f"Total validation samples: {val_gen.samples}")

# ========================================
# BUILD AND TRAIN MODEL FUNCTION
# ========================================
def build_and_train_model(model_name, base_model_class, freeze_all=True, epochs=EPOCHS):
    """
    Build, compile, and train a ResNet model
    """
    print(f"\n{'='*60}")
    print(f"ğŸ”¥ Building {model_name}")
    print(f"Status: {'FROZEN (layers not trainable)' if freeze_all else 'UNFROZEN (all layers trainable)'}")
    print(f"{'='*60}")
    
    base_model = base_model_class(
        weights="imagenet", 
        include_top=False, 
        input_shape=(IMG_SIZE[0], IMG_SIZE[1], 3)
    )
    
    x = GlobalAveragePooling2D()(base_model.output)
    x = Dropout(0.5)(x)
    output = Dense(1, activation="sigmoid")(x)
    model = Model(inputs=base_model.input, outputs=output)
    
    for layer in base_model.layers:
        layer.trainable = not freeze_all
        
    model.compile(
        optimizer=Adam(learning_rate=1e-5), 
        loss="binary_crossentropy", 
        metrics=["accuracy"]
    )
    
    total_layers = len(base_model.layers)
    trainable_count = sum([layer.trainable for layer in base_model.layers])
    frozen_count = sum([not layer.trainable for layer in base_model.layers])
    
    print(f"\nğŸ“Š Model Summary (Base Model Layers):")
    print(f"Total layers: {total_layers}")
    print(f"Trainable layers: {trainable_count}")
    print(f"Frozen layers: {frozen_count}")
    
    print(f"\nğŸš€ Training {model_name}...")
    
    steps_per_epoch = train_gen.samples // BATCH_SIZE
    validation_steps = val_gen.samples // BATCH_SIZE
    
    history = model.fit(
        train_gen,
        steps_per_epoch=steps_per_epoch,
        validation_data=val_gen,
        validation_steps=validation_steps,
        epochs=epochs,
        verbose=1
    )
    
    final_train_acc = history.history['accuracy'][-1]
    final_val_acc = history.history['val_accuracy'][-1]
    final_train_loss = history.history['loss'][-1]
    final_val_loss = history.history['val_loss'][-1]
    
    print(f"\nâœ… {model_name} Training Complete!")
    print(f"Final Training Accuracy: {final_train_acc:.4f}")
    print(f"Final Validation Accuracy: {final_val_acc:.4f}")
    print(f"Final Training Loss: {final_train_loss:.4f}")
    print(f"Final Validation Loss: {final_val_loss:.4f}")
    
    return model, history

# ========================================
# FACE DETECTION AND EXTRACTION FUNCTIONS
# ========================================
def extract_face_from_frame(frame, detector, margin=20):
    """
    Extract face region from a frame using MTCNN
    
    Args:
        frame: Input frame (RGB format)
        detector: MTCNN detector instance
        margin: Pixel margin around detected face
    
    Returns:
        Extracted face region or None if no face detected
    """
    # Detect faces
    detections = detector.detect_faces(frame)
    
    if len(detections) == 0:
        return None
    
    # Get the face with highest confidence
    best_detection = max(detections, key=lambda x: x['confidence'])
    
    # Check confidence threshold
    if best_detection['confidence'] < FACE_DETECTION_CONFIDENCE:
        return None
    
    # Extract bounding box
    x, y, w, h = best_detection['box']
    
    # Add margin
    x1 = max(0, x - margin)
    y1 = max(0, y - margin)
    x2 = min(frame.shape[1], x + w + margin)
    y2 = min(frame.shape[0], y + h + margin)
    
    # Extract face region
    face = frame[y1:y2, x1:x2]
    
    return face

def preprocess_face(face, target_size=IMG_SIZE):
    """
    Preprocess face for model input
    
    Args:
        face: Face region (RGB format)
        target_size: Target size for resizing
    
    Returns:
        Preprocessed face ready for model input
    """
    if face is None or face.size == 0:
        return None
    
    # Resize to target size
    face_resized = cv2.resize(face, target_size)
    
    return face_resized

# ========================================
# VIDEO PREDICTION FUNCTIONS
# ========================================
def extract_frames_from_video(video_path, detector, num_frames=None, frame_skip=1):
    """
    Extract frames from a video and detect faces
    
    Args:
        video_path: Path to the video file
        detector: MTCNN detector instance
        num_frames: Number of frames to extract (None = all frames)
        frame_skip: Process every Nth frame (1 = all frames, 2 = every 2nd frame, etc.)
    
    Returns:
        List of face regions as numpy arrays
    """
    faces = []
    cap = cv2.VideoCapture(str(video_path))
    
    if not cap.isOpened():
        print(f"âŒ Error: Cannot open video {video_path}")
        return faces
    
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    
    if total_frames == 0:
        print(f"âš ï¸ Warning: Video {video_path} has 0 frames")
        cap.release()
        return faces
    
    # Determine which frames to process
    if num_frames is None:
        # Extract all frames (with frame_skip)
        frame_indices = list(range(0, total_frames, frame_skip))
        print(f"   Processing ALL frames from {Path(video_path).name}...")
        print(f"   Total frames: {total_frames}, FPS: {fps}, Frame skip: {frame_skip}")
        print(f"   Frames to process: {len(frame_indices)}")
    else:
        # Extract uniformly distributed frames
        frame_indices = np.linspace(0, total_frames - 1, num_frames, dtype=int)
        print(f"   Extracting {num_frames} frames uniformly from {Path(video_path).name}...")
        print(f"   Total frames: {total_frames}, FPS: {fps}")
    
    frames_processed = 0
    faces_detected = 0
    
    for frame_idx in tqdm(frame_indices, desc=f"   Processing frames", leave=False):
        cap.set(cv2.CAP_PROP_POS_FRAMES, frame_idx)
        ret, frame = cap.read()
        
        if ret:
            frames_processed += 1
            # Convert BGR to RGB
            frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            
            # Extract face from frame
            face = extract_face_from_frame(frame_rgb, detector)
            
            if face is not None:
                # Preprocess face
                face_processed = preprocess_face(face)
                if face_processed is not None:
                    faces.append(face_processed)
                    faces_detected += 1
    
    cap.release()
    
    detection_rate = (faces_detected / frames_processed * 100) if frames_processed > 0 else 0
    print(f"   âœ“ Detected {faces_detected} faces from {frames_processed} frames ({detection_rate:.1f}%)")
    
    return faces

def predict_video(model, video_path, detector, num_frames=None, frame_skip=1):
    """
    Predict if a video is real or fake using face detection
    
    Args:
        model: Trained Keras model
        video_path: Path to the video file
        detector: MTCNN detector instance
        num_frames: Number of frames to extract (None = all frames)
        frame_skip: Process every Nth frame
    
    Returns:
        Dictionary with prediction results
    """
    # Extract faces from frames
    faces = extract_frames_from_video(video_path, detector, num_frames, frame_skip)
    
    if len(faces) == 0:
        # Get total frames for statistics
        cap = cv2.VideoCapture(str(video_path))
        total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT)) if cap.isOpened() else 0
        cap.release()
        
        return {
            'video_name': Path(video_path).name,
            'prediction': None,
            'confidence': 0.0,
            'label': 'NO_FACE_DETECTED',
            'num_faces': 0,
            'total_frames': total_frames,
            'detection_rate': 0.0
        }
    
    # Preprocess faces
    faces_array = np.array(faces) / 255.0  # Normalize to [0, 1]
    
    # Predict on all faces
    print(f"   Running prediction on {len(faces)} detected faces...")
    predictions = model.predict(faces_array, verbose=0)
    
    # Average prediction across all faces
    avg_prediction = np.mean(predictions)
    
    # Determine label (0 = fake, 1 = real)
    label = 'REAL' if avg_prediction > 0.5 else 'FAKE'
    confidence = avg_prediction if label == 'REAL' else (1 - avg_prediction)
    
    # Get total frames
    cap = cv2.VideoCapture(str(video_path))
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT)) if cap.isOpened() else 0
    cap.release()
    
    frames_to_process = total_frames if num_frames is None else num_frames
    if num_frames is None and frame_skip > 1:
        frames_to_process = total_frames // frame_skip
    
    detection_rate = len(faces) / frames_to_process if frames_to_process > 0 else 0
    
    return {
        'video_name': Path(video_path).name,
        'prediction': float(avg_prediction),
        'confidence': float(confidence),
        'label': label,
        'num_faces': len(faces),
        'total_frames': total_frames,
        'detection_rate': float(detection_rate)
    }

def predict_all_videos(model, video_dir, detector, num_frames=None, frame_skip=1):
    """
    Predict all videos in a directory using face detection
    
    Args:
        model: Trained Keras model
        video_dir: Directory containing video files
        detector: MTCNN detector instance
        num_frames: Number of frames to extract per video (None = all frames)
        frame_skip: Process every Nth frame
    
    Returns:
        DataFrame with prediction results
    """
    video_dir = Path(video_dir)
    video_files = list(video_dir.glob("*.mp4")) + list(video_dir.glob("*.avi"))
    
    if len(video_files) == 0:
        print(f"âš ï¸ No video files found in {video_dir}")
        return pd.DataFrame()
    
    print(f"\n{'='*60}")
    print(f"ğŸ¬ Found {len(video_files)} videos to process")
    if num_frames is None:
        print(f"ğŸ“Š Mode: ALL FRAMES (frame_skip={frame_skip})")
    else:
        print(f"ğŸ“Š Mode: {num_frames} frames per video (uniform sampling)")
    print(f"{'='*60}")
    
    results = []
    
    for idx, video_path in enumerate(video_files, 1):
        print(f"\n{'='*60}")
        print(f"ğŸ“¹ [{idx}/{len(video_files)}] Processing: {video_path.name}")
        print(f"{'='*60}")
        
        result = predict_video(model, video_path, detector, num_frames, frame_skip)
        results.append(result)
        
        # Print individual result
        print(f"\nğŸ“Š RESULT:")
        print(f"   Prediction: {result['label']} (confidence: {result['confidence']:.2%})")
        print(f"   Raw score: {result['prediction']:.4f}" if result['prediction'] is not None else "   Raw score: N/A")
        print(f"   Faces detected: {result['num_faces']} (rate: {result['detection_rate']:.1%})")
        print(f"   Total frames in video: {result['total_frames']}")
    
    return pd.DataFrame(results)

# ========================================
# IMAGE PREDICTION FUNCTIONS
# ========================================
def predict_image(model, image_path, detector):
    """
    Predict if an image is real or fake using face detection
    
    Args:
        model: Trained Keras model
        image_path: Path to the image file
        detector: MTCNN detector instance
    
    Returns:
        Dictionary with prediction results
    """
    # Read image
    image = cv2.imread(str(image_path))
    
    if image is None:
        return {
            'image_name': Path(image_path).name,
            'prediction': None,
            'confidence': 0.0,
            'label': 'ERROR_READING_IMAGE',
            'face_detected': False
        }
    
    # Convert BGR to RGB
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    # Extract face from image
    face = extract_face_from_frame(image_rgb, detector)
    
    if face is None:
        return {
            'image_name': Path(image_path).name,
            'prediction': None,
            'confidence': 0.0,
            'label': 'NO_FACE_DETECTED',
            'face_detected': False
        }
    
    # Preprocess face
    face_processed = preprocess_face(face)
    
    if face_processed is None:
        return {
            'image_name': Path(image_path).name,
            'prediction': None,
            'confidence': 0.0,
            'label': 'PREPROCESSING_ERROR',
            'face_detected': True
        }
    
    # Normalize and add batch dimension
    face_array = np.expand_dims(face_processed / 255.0, axis=0)
    
    # Predict
    prediction = model.predict(face_array, verbose=0)[0][0]
    
    # Determine label (0 = fake, 1 = real)
    label = 'REAL' if prediction > 0.5 else 'FAKE'
    confidence = prediction if label == 'REAL' else (1 - prediction)
    
    return {
        'image_name': Path(image_path).name,
        'prediction': float(prediction),
        'confidence': float(confidence),
        'label': label,
        'face_detected': True
    }

def predict_all_images(model, image_dir, detector):
    """
    Predict all images in a directory using face detection
    
    Args:
        model: Trained Keras model
        image_dir: Directory containing image files
        detector: MTCNN detector instance
    
    Returns:
        DataFrame with prediction results
    """
    image_dir = Path(image_dir)
    image_extensions = ['*.jpg', '*.jpeg', '*.png', '*.bmp']
    image_files = []
    for ext in image_extensions:
        image_files.extend(list(image_dir.glob(ext)))
    
    if len(image_files) == 0:
        print(f"âš ï¸ No image files found in {image_dir}")
        return pd.DataFrame()
    
    print(f"\n{'='*60}")
    print(f"ğŸ–¼ï¸ Found {len(image_files)} images to process")
    print(f"{'='*60}")
    
    results = []
    
    for image_path in tqdm(image_files, desc="Processing images"):
        result = predict_image(model, image_path, detector)
        results.append(result)
        
        # Print individual result
        if result['face_detected']:
            print(f"\nğŸ“· {result['image_name']}")
            print(f"   Prediction: {result['label']} (confidence: {result['confidence']:.2%})")
            print(f"   Raw score: {result['prediction']:.4f}" if result['prediction'] is not None else "   Raw score: N/A")
        else:
            print(f"\nğŸ“· {result['image_name']}: {result['label']}")
    
    return pd.DataFrame(results)

# ========================================
# TRAIN ALL MODELS
# ========================================
models_results = {}

# 1. ResNet50 - Frozen (Transfer Learning)
model_r50_frozen, history_r50_frozen = build_and_train_model(
    "ResNet50 Frozen", 
    ResNet50, 
    freeze_all=True
)
models_results['ResNet50_Frozen'] = {
    'model': model_r50_frozen, 
    'history': history_r50_frozen
}

# 2. ResNet50 - Unfrozen (Fine Tuning)
model_r50_unfrozen, history_r50_unfrozen = build_and_train_model(
    "ResNet50 Unfrozen", 
    ResNet50, 
    freeze_all=False
)
models_results['ResNet50_Unfrozen'] = {
    'model': model_r50_unfrozen, 
    'history': history_r50_unfrozen
}

# 3. ResNet50V2 - Frozen (Transfer Learning)
model_r50v2_frozen, history_r50v2_frozen = build_and_train_model(
    "ResNet50V2 Frozen", 
    ResNet50V2, 
    freeze_all=True
)
models_results['ResNet50V2_Frozen'] = {
    'model': model_r50v2_frozen, 
    'history': history_r50v2_frozen
}

# 4. ResNet50V2 - Unfrozen (Fine Tuning)
model_r50v2_unfrozen, history_r50v2_unfrozen = build_and_train_model(
    "ResNet50V2 Unfrozen", 
    ResNet50V2, 
    freeze_all=False
)
models_results['ResNet50V2_Unfrozen'] = {
    'model': model_r50v2_unfrozen, 
    'history': history_r50v2_unfrozen
}

# ========================================
# EVALUATION - COMPARE ALL MODELS
# ========================================
print(f"\n{'='*70}")
print("ğŸ“ˆ EVALUATION - COMPARISON OF ALL MODELS")
print(f"{'='*70}")
print(f"{'Model':<35} {'Train Acc':<12} {'Val Acc':<12} {'Val Loss':<12}")
print(f"{'-'*70}")

evaluation_results = []
for model_name, result in models_results.items():
    history = result['history']
    train_acc = history.history['accuracy'][-1]
    val_acc = history.history['val_accuracy'][-1]
    val_loss = history.history['val_loss'][-1]
    
    evaluation_results.append({
        'Model': model_name,
        'Train_Acc': train_acc,
        'Val_Acc': val_acc,
        'Val_Loss': val_loss
    })
    
    print(f"{model_name:<35} {train_acc:<12.4f} {val_acc:<12.4f} {val_loss:<12.4f}")

eval_df = pd.DataFrame(evaluation_results)

best_model_name = max(models_results.keys(), 
                      key=lambda x: models_results[x]['history'].history['val_accuracy'][-1])
best_model = models_results[best_model_name]['model']
best_val_acc = models_results[best_model_name]['history'].history['val_accuracy'][-1]

print(f"\n{'='*70}")
print(f"ğŸ† BEST MODEL: {best_model_name}")
print(f"ğŸ¯ Best Validation Accuracy: {best_val_acc:.4f}")
print(f"{'='*70}")

# ========================================
# SAVE MODELS
# ========================================
print(f"\n{'='*60}")
print("ğŸ’¾ Saving Models...")
print(f"{'='*60}")

save_dir = Path("/kaggle/working/resnet_models")

if save_dir.exists():
    shutil.rmtree(save_dir)
save_dir.mkdir(exist_ok=True)

for model_name, result in models_results.items():
    save_path = save_dir / f"{model_name.lower()}_model.h5"
    try:
        result['model'].save(str(save_path))
        print(f"âœ… Saved: {save_path}")
    except Exception as e:
        print(f"âŒ Error saving {model_name}: {e}")

eval_csv_path = save_dir / "evaluation_results.csv"
eval_df.to_csv(eval_csv_path, index=False)
print(f"\nâœ… Evaluation results saved to: {eval_csv_path}")

# ========================================
# VIDEO PREDICTION ON TEST DATA
# ========================================
print(f"\n{'='*70}")
print("ğŸ¬ VIDEO PREDICTION ON TEST DATA")
print(f"{'='*70}")

# Predict with best model
print(f"\nğŸ”® Using best model: {best_model_name}")
video_predictions = predict_all_videos(best_model, TEST_VIDEO_DIR, detector, 
                                      num_frames=FRAMES_PER_VIDEO, 
                                      frame_skip=FRAME_SKIP)

if not video_predictions.empty:
    # Save predictions
    predictions_csv_path = save_dir / "video_predictions.csv"
    video_predictions.to_csv(predictions_csv_path, index=False)
    print(f"\nâœ… Video predictions saved to: {predictions_csv_path}")
    
    # Summary statistics
    print(f"\n{'='*60}")
    print("ğŸ“Š VIDEO PREDICTION SUMMARY")
    print(f"{'='*60}")
    print(f"Total videos processed: {len(video_predictions)}")
    print(f"Predicted REAL: {(video_predictions['label'] == 'REAL').sum()}")
    print(f"Predicted FAKE: {(video_predictions['label'] == 'FAKE').sum()}")
    print(f"No face detected: {(video_predictions['label'] == 'NO_FACE_DETECTED').sum()}")
    
    # Calculate statistics only for videos with detected faces
    valid_predictions = video_predictions[video_predictions['label'].isin(['REAL', 'FAKE'])]
    if not valid_predictions.empty:
        print(f"Average confidence (valid predictions): {valid_predictions['confidence'].mean():.2%}")
        print(f"Average face detection rate: {valid_predictions['detection_rate'].mean():.1%}")
    
    print(f"\nDetailed Results:")
    print(video_predictions.to_string(index=False))
else:
    print("\nâš ï¸ No videos were processed")

# ========================================
# IMAGE PREDICTION ON TEST DATA
# ========================================
print(f"\n{'='*70}")
print("ğŸ–¼ï¸ IMAGE PREDICTION ON TEST DATA")
print(f"{'='*70}")

# Check if test image directory exists
if TEST_IMAGE_DIR.exists():
    print(f"\nğŸ”® Using best model: {best_model_name}")
    image_predictions = predict_all_images(best_model, TEST_IMAGE_DIR, detector)
    
    if not image_predictions.empty:
        # Save predictions
        image_predictions_csv_path = save_dir / "image_predictions.csv"
        image_predictions.to_csv(image_predictions_csv_path, index=False)
        print(f"\nâœ… Image predictions saved to: {image_predictions_csv_path}")
        
        # Summary statistics
        print(f"\n{'='*60}")
        print("ğŸ“Š IMAGE PREDICTION SUMMARY")
        print(f"{'='*60}")
        print(f"Total images processed: {len(image_predictions)}")
        print(f"Predicted REAL: {(image_predictions['label'] == 'REAL').sum()}")
        print(f"Predicted FAKE: {(image_predictions['label'] == 'FAKE').sum()}")
        print(f"No face detected: {(image_predictions['label'] == 'NO_FACE_DETECTED').sum()}")
        print(f"Errors: {(image_predictions['label'].str.contains('ERROR')).sum()}")
        
        # Calculate statistics only for images with detected faces
        valid_predictions = image_predictions[image_predictions['label'].isin(['REAL', 'FAKE'])]
        if not valid_predictions.empty:
            print(f"Average confidence (valid predictions): {valid_predictions['confidence'].mean():.2%}")
            print(f"Face detection success rate: {(image_predictions['face_detected']).sum() / len(image_predictions):.1%}")
        
        print(f"\nDetailed Results:")
        print(image_predictions.to_string(index=False))
    else:
        print("\nâš ï¸ No images were processed")
else:
    print(f"\nâš ï¸ Test image directory not found: {TEST_IMAGE_DIR}")
    print("   Skipping image prediction...")

# ========================================
# DETAILED TRAINING HISTORY
# ========================================
print(f"\n{'='*60}")
print("ğŸ“Š DETAILED TRAINING HISTORY")
print(f"{'='*60}")

for model_name, result in models_results.items():
    history = result['history']
    print(f"\n{model_name}:")
    print(f"  Epoch | Train Acc | Val Acc    | Train Loss | Val Loss")
    print(f"  {'-'*60}")
    for epoch in range(len(history.history['accuracy'])):
        print(f"  {epoch+1:<5} | {history.history['accuracy'][epoch]:.4f}    | "
              f"{history.history['val_accuracy'][epoch]:.4f}    | "
              f"{history.history['loss'][epoch]:.4f}      | "
              f"{history.history['val_loss'][epoch]:.4f}")

print("\n" + "="*60)
print("âœ… ALL TRAINING AND EVALUATION COMPLETE!")
print("="*60)

# ========================================
# PLOT TRAINING HISTORY
# ========================================
print(f"\n{'='*60}")
print("ğŸ“Š GENERATING TRAINING PLOTS...")
print(f"{'='*60}")

fig, axes = plt.subplots(4, 2, figsize=(16, 20))
fig.suptitle('Training History - All Models', fontsize=20, fontweight='bold', y=0.995)

model_names = list(models_results.keys())

for idx, model_name in enumerate(model_names):
    history = models_results[model_name]['history']
    
    ax_acc = axes[idx, 0]
    ax_acc.plot(history.history['accuracy'], label='Train Accuracy', linewidth=2, marker='o')
    ax_acc.plot(history.history['val_accuracy'], label='Val Accuracy', linewidth=2, marker='s')
    ax_acc.set_title(f'{model_name} - Accuracy', fontsize=14, fontweight='bold')
    ax_acc.set_xlabel('Epoch', fontsize=12)
    ax_acc.set_ylabel('Accuracy', fontsize=12)
    ax_acc.legend(loc='lower right', fontsize=10)
    ax_acc.grid(True, alpha=0.3)
    ax_acc.set_ylim([0, 1])
    
    ax_loss = axes[idx, 1]
    ax_loss.plot(history.history['loss'], label='Train Loss', linewidth=2, marker='o', color='coral')
    ax_loss.plot(history.history['val_loss'], label='Val Loss', linewidth=2, marker='s', color='red')
    ax_loss.set_title(f'{model_name} - Loss', fontsize=14, fontweight='bold')
    ax_loss.set_xlabel('Epoch', fontsize=12)
    ax_loss.set_ylabel('Loss', fontsize=12)
    ax_loss.legend(loc='upper right', fontsize=10)
    ax_loss.grid(True, alpha=0.3)

plt.tight_layout()
plot_path = save_dir / "training_history_all_models.png"
plt.savefig(plot_path, dpi=300, bbox_inches='tight')
print(f"âœ… Saved combined plot: {plot_path}")

# ========================================
# PLOT COMPARISON - FROZEN VS UNFROZEN
# ========================================
print(f"\n{'='*60}")
print("ğŸ“Š GENERATING COMPARISON PLOTS...")
print(f"{'='*60}")

fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Comparison: Frozen vs Unfrozen Models', fontsize=20, fontweight='bold')

ax1 = axes[0, 0]
ax1.plot(models_results['ResNet50_Frozen']['history'].history['val_accuracy'], 
          label='ResNet50 Frozen', linewidth=2, marker='o')
ax1.plot(models_results['ResNet50_Unfrozen']['history'].history['val_accuracy'], 
          label='ResNet50 Unfrozen', linewidth=2, marker='s')
ax1.set_title('ResNet50 - Validation Accuracy', fontsize=14, fontweight='bold')
ax1.set_xlabel('Epoch', fontsize=12)
ax1.set_ylabel('Validation Accuracy', fontsize=12)
ax1.legend(fontsize=10)
ax1.grid(True, alpha=0.3)
ax1.set_ylim([0, 1])

ax2 = axes[0, 1]
ax2.plot(models_results['ResNet50_Frozen']['history'].history['val_loss'], 
          label='ResNet50 Frozen', linewidth=2, marker='o', color='coral')
ax2.plot(models_results['ResNet50_Unfrozen']['history'].history['val_loss'], 
          label='ResNet50 Unfrozen', linewidth=2, marker='s', color='red')
ax2.set_title('ResNet50 - Validation Loss', fontsize=14, fontweight='bold')
ax2.set_xlabel('Epoch', fontsize=12)
ax2.set_ylabel('Validation Loss', fontsize=12)
ax2.legend(fontsize=10)
ax2.grid(True, alpha=0.3)

ax3 = axes[1, 0]
ax3.plot(models_results['ResNet50V2_Frozen']['history'].history['val_accuracy'], 
          label='ResNet50V2 Frozen', linewidth=2, marker='o')
ax3.plot(models_results['ResNet50V2_Unfrozen']['history'].history['val_accuracy'], 
          label='ResNet50V2 Unfrozen', linewidth=2, marker='s')
ax3.set_title('ResNet50V2 - Validation Accuracy', fontsize=14, fontweight='bold')
ax3.set_xlabel('Epoch', fontsize=12)
ax3.set_ylabel('Validation Accuracy', fontsize=12)
ax3.legend(fontsize=10)
ax3.grid(True, alpha=0.3)
ax3.set_ylim([0, 1])

ax4 = axes[1, 1]
ax4.plot(models_results['ResNet50V2_Frozen']['history'].history['val_loss'], 
          label='ResNet50V2 Frozen', linewidth=2, marker='o', color='coral')
ax4.plot(models_results['ResNet50V2_Unfrozen']['history'].history['val_loss'], 
          label='ResNet50V2 Unfrozen', linewidth=2, marker='s', color='red')
ax4.set_title('ResNet50V2 - Validation Loss', fontsize=14, fontweight='bold')
ax4.set_xlabel('Epoch', fontsize=12)
ax4.set_ylabel('Validation Loss', fontsize=12)
ax4.legend(fontsize=10)
ax4.grid(True, alpha=0.3)

plt.tight_layout()
comparison_plot_path = save_dir / "comparison_frozen_vs_unfrozen.png"
plt.savefig(comparison_plot_path, dpi=300, bbox_inches='tight')
print(f"âœ… Saved comparison plot: {comparison_plot_path}")

print("\n" + "="*60)
print("ğŸ‰ ALL PLOTS GENERATED SUCCESSFULLY!")
print("="*60)
```
